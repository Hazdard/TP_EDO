$\textbf{Remarque :}$ Dans les portraits de phase ci-dessous, les trajectoires sont en bleu et l'isocline horizontale (resp. verticale) est en vert (resp. en rouge). On a de plus tracé le segment [(0,2),(2,0)] en violet.

```python
#Question 1 et 2

for mu in [1,2.5] :

    f=lambda y,t: np.array([y[0]*y[1]-mu*y[0],-y[0]*y[1] + mu*(2-y[1])])

    xmin=0
    xmax=2.5
    ymin=0
    ymax=3

    ti=0
    tf=8
    h=0.02
    T=np.arange(ti,tf+h,h)

    CI= rd.rand(15,2)*3

    for y0 in CI :
            Sode=odeint(f,y0,T)
            plt.plot(Sode[:,0],Sode[:,1],color="blue")

    plt.axis([xmin,xmax,ymin,ymax])
    envpp(f,xmin,xmax,ymin,ymax)
    plt.plot([0,2],[2,0], color="purple",linewidth=3)
    plt.figure()
```
