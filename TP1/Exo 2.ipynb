$\textbf{Remarque :}$ Dans les portraits de phase ci-dessous, les trajectoires sont en bleu, l'isocline horizontale (resp. verticale) est en vert (resp. en rouge), tandis que les droites dirigées par les vecteurs propres sont en violet et en gras (quand elles existent).

```python
#Question 1 et 2

for a,b,c,d in tab_abcd :

    f=lambda y,t: np.array([a*y[0]+b*y[1],c*y[0]+d*y[1]])

    xmin=-4
    xmax=4
    ymin=-4
    ymax=4

    ti=0
    tf=8
    h=0.01
    T=np.arange(ti,tf+h,h)

    CI= (rd.rand(15,2)-0.5)*4

    for y0 in CI :
            Sode=odeint(f,y0,T)
            Sode_neg=odeint(f,y0,-T)
            plt.plot(Sode[:,0],Sode[:,1],color="blue")
            plt.plot(Sode_neg[:,0],Sode_neg[:,1], color="blue")

    plt.axis([xmin,xmax,ymin,ymax])
    envpp(f,xmin,xmax,ymin,ymax)
 
    #Question 3

    A= np.matrix([[a,b],[c,d]])
    D,V=eig(A)
    print("Valeurs propres :", D,"\n")
    print("Vecteurs propres : \n", V)
    
    plt.xlabel("Composante x")
    plt.ylabel("Composante y")
    plt.title(' v = {:9.1f}'.format(D[0]))

    #Question 4

    if D[0].imag == 0 :
        plt.plot([-6*V[0][0],6*V[0][0]],[-6*V[1][0],6*V[1][0]], color="purple",linewidth=3)
        plt.plot([-6*V[0][1],6*V[0][1]],[-6*V[1][1],6*V[1][1]], color="purple",linewidth=3)
    plt.figure()
```
