$\textbf{Remarque :}$ Dans les portraits de phase ci-dessous, les trajectoires sont en bleu tandis que l'isoclines horizontale est en vert et l'isocline verticale est en rouge. 

```python
a=1

f=lambda y,t: np.array([y[1]-a*(y[0]**3 - y[0]),-y[0]])

xmin=-1
xmax=1
ymin=-1
ymax=1

ti=0
tf=10*a +20
h=0.01
T=np.arange(ti,tf+h,h)

CI= (rd.rand(20,2)-0.5)*2

for y0 in CI :
        Sode=odeint(f,y0,T)
        plt.plot(Sode[:,0],Sode[:,1], color="blue")

plt.axis([xmin,xmax,ymin,ymax])
envpp(f,xmin,xmax,ymin,ymax)

plt.xlabel("Composante x")
plt.ylabel("Composante y")
plt.title("Portrait de phase")
```

```python
#Question 2 et 3

tab_a=[1,20] 

for a in tab_a :
    tf=10*a +20
    T=np.arange(ti,tf,h)
    y0=[0.0,0.1]
    Sode, out=odeint(f,y0,T, full_output=True)
    plt.subplot(3,1,1)
    plt.title("Composante x en fonction du temps, a=%i"%a)
    plt.ylabel("Valeur de x(t)")
    plt.plot(T,Sode[:,0])
    plt.subplot(3,1,2)
    plt.title("Composante y en fonction du temps, a=%i"%a)
    plt.ylabel("Valeur de y(t)")
    plt.plot(T,Sode[:,1])
    plt.subplot(3,1,3)
    plt.title("Temps réel en fonction du temps, a=%i"%a)
    plt.ylabel("Pas de temps réel")
    plt.semilogy(T[1:],out['hu'])
 
    plt.figure()
```
